### **Посчитайте Retention 15 дня (в процентах) для пользователей, зарегистрированных в январе**

Cохраните результат в переменную `retention_15_day`

**Примечание:** результат округлите до 5 знаков после запятой

```python
import csv
from datetime import datetime

registration_dates = {}
with open('registrations.csv', 'r') as f1:
    reader = csv.reader(f1, delimiter=';')
    next(reader)
    for row in reader:
        user_id = row[0]
        registration_date = row[1]
        registration_dates[user_id] = registration_date

#unique_users=set( registration_dates.keys())

combined_data = []
with open('entries.csv', 'r') as f2:
    reader = csv.reader(f2, delimiter=';')
    next(reader)
    for row in reader:
        user_id = row[0]
        entry_date_str = row[1]
        reg_date_str = registration_dates.get(user_id)

        if reg_date_str:
            reg_date = datetime.strptime(reg_date_str, '%Y-%m-%d')
            login_date = datetime.strptime(entry_date_str, '%Y-%m-%d')

            combined_data.append([
                user_id,
                reg_date,
                login_date
            ])

filter_date = datetime.strptime('2021-02-01', '%Y-%m-%d')
users= set()
day15_users= set()
for row in combined_data:
  user_id=row[0]
  reg_date=row[1]
  login_date=row[2]
  delta=(login_date - reg_date).days
  if reg_date < filter_date:
     users.add(user_id)
     if delta == 15:
      day15_users.add(user_id)

retention_15_day = round(len(day15_users)/len(users)*100,5)

print(retention_15_day)


```

### **Посчитайте Lifetime по всем пользователям, посчитанный как интеграл от n-day retention**

Сохраните результат в переменную `lifetime`

**Примечание:** результат округлите до 5 знаков после запятой

```python
import csv
from datetime import datetime

from collections import defaultdict

registration_dates = {}
with open('registrations.csv', 'r') as f1:
    reader = csv.reader(f1, delimiter=';')
    next(reader)
    for row in reader:
        user_id = row[0]
        registration_date = row[1]
        registration_dates[user_id] = registration_date

#unique_users=set( registration_dates.keys())

combined_data = []
with open('entries.csv', 'r') as f2:
    reader = csv.reader(f2, delimiter=';')
    next(reader)
    for row in reader:
        user_id = row[0]
        entry_date_str = row[1]
        reg_date_str = registration_dates.get(user_id)

        if reg_date_str:
            reg_date = datetime.strptime(reg_date_str, '%Y-%m-%d')
            login_date = datetime.strptime(entry_date_str, '%Y-%m-%d')

            combined_data.append({
                'user_id': user_id,
                'registration_date': reg_date,
                'login_date': login_date
            })

day_users = defaultdict(set)
for row in combined_data:
    n_day = (row['login_date'] - row['registration_date']).days
    day_users[n_day].add(row['user_id'])


total_users = len(registration_dates)  # общее количество уникальных пользователей
retention = {}
for n_day, users in day_users.items():
    retention[n_day] = len(users) / total_users


lifetime = round(sum(retention.values()),5)

print(lifetime)





```

### **Посчитайте Churn rate 29 дня (в долях), посчитанный по всем пользователям**

Сохраните результат в переменную `churn_29`.  
Если будете считать CR от Retention, ведите расчет от Rolling Retention, таким образом, мы получим всех, кто не заходил в 29 день и после. Ведя расчет от обычного Retention, наоборот - получим только CR в 29 день.


```python
import csv
from datetime import datetime

registration_dates = {}
with open('registrations.csv', 'r') as f1:
    reader = csv.reader(f1, delimiter=';')
    next(reader)
    for row in reader:
        user_id = row[0]
        registration_date = row[1]
        registration_dates[user_id] = registration_date

#unique_users=set( registration_dates.keys())

combined_data = []
with open('entries.csv', 'r') as f2:
    reader = csv.reader(f2, delimiter=';')
    next(reader)
    for row in reader:
        user_id = row[0]
        entry_date_str = row[1]
        reg_date_str = registration_dates.get(user_id)

        if reg_date_str:
            reg_date = datetime.strptime(reg_date_str, '%Y-%m-%d')
            login_date = datetime.strptime(entry_date_str, '%Y-%m-%d')

            combined_data.append([
                user_id,
                reg_date,
                login_date
            ])

users= set()
day29_users= set()
for row in combined_data:
  user_id=row[0]
  reg_date=row[1]
  login_date=row[2]
  delta=(login_date - reg_date).days
  users.add(user_id)
  if delta >= 29:
    day29_users.add(user_id)

rolling_retention = round(len(day29_users)/len(users),5)
churn_29 = round((1- rolling_retention),5)

print(churn_29)


```

### **Посчитайте Mau, Wau, Dau усредненные**

Сохраните результат в переменные `avg_mau`, `avg_wau`, `avg_dau` соответственно

**Примечание:** результаты округлите до 5 знаков после запятой

```python
import csv
from datetime import datetime

from collections import defaultdict

registration_dates = {}
with open('registrations.csv', 'r') as f1:
    reader = csv.reader(f1, delimiter=';')
    next(reader)
    for row in reader:
        user_id = row[0]
        registration_date = row[1]
        registration_dates[user_id] = registration_date

#unique_users=set( registration_dates.keys())

combined_data = []
with open('entries.csv', 'r') as f2:
    reader = csv.reader(f2, delimiter=';')
    next(reader)
    for row in reader:
        user_id = row[0]
        entry_date_str = row[1]
        reg_date_str = registration_dates.get(user_id)

        if reg_date_str:
            reg_date = datetime.strptime(reg_date_str, '%Y-%m-%d')
            login_date = datetime.strptime(entry_date_str, '%Y-%m-%d')

            combined_data.append({
                'user_id': user_id,
                'registration_date': reg_date,
                'login_date': login_date
            })



month_users = defaultdict(set)
for row in combined_data:
  login_date = row['login_date']
  n_month = login_date.strftime('%Y-%m')
  month_users[n_month].add(row['user_id'])

  maus = {}
  for month, users in month_users.items():
    maus[month] = len(users)

for month, mau in sorted(maus.items()):
  avg_mau = round((sum(maus.values()) / len(maus)),5)

print(avg_mau)
```
