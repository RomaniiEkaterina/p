### **Найдите самую эффективную платформу по средним значениям конверсий**

Cохраните результат в переменную `most_effective_platform`

````python
import csv
from collections import defaultdict

Platform_col = "Платформа"
conv_col = "Конверсия"

sum_by_platform = defaultdict(float)
count_by_platform = defaultdict(int)

with open("campaign_data.csv", "r", encoding="utf-8-sig", newline="") as f:
    reader = csv.DictReader(f)
    for row in reader:
        platform = (row.get(Platform_col) or "").strip()
        conv_str = (row.get(conv_col) or "").strip()

        if conv_str:
            try:
                value = float(conv_str.replace(",", "."))
                sum_by_platform[platform] += value
                count_by_platform[platform] += 1
            except ValueError:
                pass

avg_by_platform = {
    p: (sum_by_platform[p] / count_by_platform[p])
    for p in sum_by_platform
    if count_by_platform[p] > 0
}


if avg_by_platform:
    most_effective_platform = max(avg_by_platform, key=avg_by_platform.get)
    print(most_effective_platform)
````

### **Посчитайте долю затрат на рекламные кампании каждой платформы и каждого города**

Сохраните результат в файл `platform_city_results.txt`

Результат должен иметь такой же порядок групп, что и изначальный файл. А города внутри платформ должны быть отсортированы по убыванию доли затрат, а доля затрат должны быть округлена до 2 знаков после запятой.

Результат должен выглядеть примерно так:

```
Для группы Telegram:
- Город: Москва, доля затрат на рекламу: 16.46%
- Город: Калуга, доля затрат на рекламу: 16.2%
- Город: Казань, доля затрат на рекламу: 14.57%
- ...
Для группы Вконтакте:
- Город: Екатеринбург, доля затрат на рекламу: 17.43%
- Город: Новосибирск, доля затрат на рекламу: 14.86%
- Город: Санкт-Петербург, доля затрат на рекламу: 14.13%
...
```

````python
import csv
from collections import defaultdict

Platform_col = "Платформа"
city_col = "Город"
budget_col ="Бюджет"

sum_budget_pl = defaultdict(float)
sum_budget_c = defaultdict(float)
platform_order = []


with open("campaign_data.csv", "r", encoding="utf-8-sig", newline="") as f:
    reader = csv.DictReader(f)
    for row in reader:
        platform = (row.get(Platform_col) or "").strip()
        city  = (row.get(city_col ) or "").strip()
        budget = (row.get(budget_col) or "").strip()
        if not budget:
          continue
        try:
          budget_value = float(budget.replace(",", "."))
        except ValueError:
           continue

        if platform:
              sum_budget_pl[platform] += budget_value

        if platform and city:
                key = (platform, city)
                sum_budget_c[key] += budget_value
                if platform not in platform_order:
                    platform_order.append(platform)


avg_by_platform_city = {
    key: round((sum_budget_c[key] / sum_budget_pl[key[0]])*100,2)
    for key in sum_budget_c
    if sum_budget_pl[key[0]] > 0
}


grouped = defaultdict(list)
for (platform, city), avg in avg_by_platform_city.items():
    grouped[platform].append((city, round(avg, 2)))

for platform in grouped:
    grouped[platform].sort(key=lambda x: x[1], reverse=True)


output_file = "platform_city_results.txt"
with open(output_file, "w", encoding="utf-8") as f:
    for platform in platform_order:
        if platform in grouped:
            f.write(f"Для группы {platform}:\n")
            for city, value in grouped[platform]:
                f.write(f"- Город: {city}, доля затрат на рекламу: {value}%\n")


with open("platform_city_results.txt", "r", encoding="utf-8") as f:
    content = f.read()
    print(content)
````

### **Напишите функцию, которая возвращает средний бюджет всех кампаний, запущенных на определенной платформе**

Функция `calculate_average_budget` должна принимать путь к файлу и название плафтормы и возвращать число - средний бюджет всех кампаний, запущенных на данной платформе, округленный до двух знаков после запятой.

````python
import csv

def calculate_average_budget(file_path, platform_name):
    total_budget = 0.0
    count = 0

    with open(file_path, "r", encoding="utf-8-sig", newline="") as f:
        reader = csv.DictReader(f, delimiter=",")
        for row in reader:
            platform = (row.get("Платформа") or "").strip()
            budget_str = (row.get("Бюджет") or "").strip()

            if platform.lower() == platform_name.lower() and budget_str:
                try:
                    budget = float(budget_str.replace(",", "."))
                    total_budget += budget
                    count += 1
                except ValueError:
                    pass


    average = total_budget / count
    return round(average, 2)

calculate_average_budget('campaign_data.csv', 'Google')
````

### **Напишите генератор, который будет генерировать даты, в которые НЕ запускались кампании**

Функция `get_missing_campaign_dates` должна принимать путь к файлу и генерировать построчно даты, отсортированные по возрастанию, в которые не запускались кампании, в формате:

```
2022-02-29
2022-03-28
2022-03-29
2022-05-30
2022-06-13
2022-06-27
...
```

````python

import csv
from datetime import datetime, timedelta

def get_missing_campaign_dates(file_path):
    dates = []

    with open(file_path, "r", encoding="utf-8-sig", newline="") as f:
        reader = csv.DictReader(f, delimiter=",")
        for row in reader:
            date_str = (row.get("Начальная дата") or "").strip()
            if date_str:
                try:
                    date = datetime.strptime(date_str, "%Y-%m-%d").date()
                    dates.append(date)
                except ValueError:
                    print(f" Некорректный формат даты: {date_str}")

    if not dates:
        return


    dates = sorted(dates)
    for i in range(len(dates) - 1):
        current = dates[i]
        next_date = dates[i + 1]
        delta = (next_date - current).days
        if delta > 1:
            for j in range(1, delta):
                missing_date = current + timedelta(days=j)
                yield missing_date

for date in get_missing_campaign_dates('campaign_data.csv'):
    print(date)
````

### **Напишите функцию, которая будет принимать на вход путь к файлу и возвращать сгруппированные данные по городам, содержащие суммарный бюджет и количество кликов**

Функция `group_campaign_data` должна возвращать список словарей, отсортированный по городам, в формате:

```python
[{'Город': 'Нижний Новгород',
  'Количество кликов': 616.0,
  'Суммарный бюджет': 27585},
 {'Город': 'Новосибирск',
  'Количество кликов': 733.0,
  'Суммарный бюджет': 33001},
  ...]
```

````python
import csv
from collections import defaultdict

def group_campaign_data(file_path):
    click_col = "Клики"
    city_col = "Город"
    budget_col = "Бюджет"

    sum_click = defaultdict(float)
    sum_budget = defaultdict(float)

    with open(file_path, "r", encoding="utf-8-sig", newline="") as f:
        reader = csv.DictReader(f)
        for row in reader:
            city_str = (row.get(city_col) or "").strip()
            budget = (row.get(budget_col) or "").strip()
            click = (row.get(click_col) or "").strip()

            if not city or not budget or not click:
                continue

            try:
                budget_value = float(budget.replace(",", "."))
                click_value = float(click.replace(",", "."))
            except ValueError:
                continue

            sum_budget[city] += budget_value
            sum_click[city] += click_value


    result = [
        {
            "Город": city,
            "Количество кликов": round(sum_click[city], 1),
            "Суммарный бюджет": int(sum_budget[city]),
        }
        for city in sorted(sum_budget.keys())
    ]

    return result

    data = group_campaign_data("campaign_data.csv")
    for row in data:
        print(row)
````

### **Напишите генератор, который будет генерировать данные о кампаниях, запущенных в указанном городе и имеющих бюджет выше заданного значения**

Функция `campaign_generator` должна принимать путь к файлу, город и бюджет и генерировать словари, отсортированные по возрастанию 'ID Кампании' в формате:

```
{'ID Кампании': '657', 'Тип': 'Контекстная', 'Платформа': 'Вконтакте', 'Доход': '10692'}
{'ID Кампании': '674', 'Тип': 'Нативная интеграция', 'Платформа': 'Яндекс', 'Доход': '20766'}
{'ID Кампании': '761', 'Тип': 'Соцсети', 'Платформа': 'Вконтакте', 'Доход': '15680'}
```

````python
import csv

def campaign_generator(file_path, city, min_budget):
    campaigns = []

    with open(file_path, "r", encoding="utf-8-sig", newline="") as f:
        reader = csv.DictReader(f, delimiter=",")
        for row in reader:
            try:
                budget = float(row.get("Бюджет", 0))
            except ValueError:
                continue

            if row.get("Город") == city and budget > min_budget:
                campaigns.append(row)


    campaigns.sort(key=lambda x: int(x.get("ID Кампании", 0)))


    for camp in campaigns:
        yield {
            "ID Кампании": camp["ID Кампании"],
            "Тип": camp["Тип кампании"],
            "Платформа": camp["Платформа"],
            "Доход": camp["Доход"]
        }


for campaign in campaign_generator('campaign_data.csv', 'Москва', 25000):
    print(campaign)
````
